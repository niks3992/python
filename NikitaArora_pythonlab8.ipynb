1. Write a python program to read an image and do the following:
• Get the information related to the image.
• Rotate the image to specific angle given by user.
• Flip the image to left-right and top-bottom.
• Resize the image to new size.
• Thumbnail of your image.
• convert image to grey scale and black-white.
• Apply different filters on the image.

```python
from PIL import Image
```

```python
img=Image.open("deer.jpeg")
print("Size of image is",img.size)
print("Format of image is",img.format)
print("File name is",img.filename)
print("Height of image is",img.height)
print("Width of image is",img.width)
img
```

Rotate the image to specific angle given by user.

```python
angle=int(input("Enter the angle"))
img1=img.rotate(angle,expand=1)
img1
```

Flip the image to left-right and top-bottom.

```python
ver_img=img.transpose(method=Image.FLIP_TOP_BOTTOM)
ver_img
```

```python
hor_img=img.transpose(method=Image.FLIP_TOP_BOTTOM)
hor_img
```

Resize the image to new size

```python
size1=eval(input("Enter the new size in tuple"))
r_resize=img.resize(size1)
r_resize
```

convert image to grey scale and black-white

```python
img3=img.convert('L')
img3
```

```python
img4=img.convert('1')
img4
```

 Apply different filters on the image

```python
from PIL import ImageFilter
```

```python
img5=img.filter(ImageFilter.BLUR)
img5
```

```python
img6=img.filter(ImageFilter.BoxBlur(30))
img6
```

```python
img6=img.filter(ImageFilter.GaussianBlur(2))
img6
```

Thumbnail of your image.

```python
size2=eval(input("Enter the new size in tuple for thumbnail"))
img.thumbnail(size2)
img
```

4. Write a python program to read a url and do the following:
● Get the title of the webpage.
● Get the body of the webpage and create a text file with it.
● Get all the urls that the web page refers to.

```python
import requests
from bs4 import BeautifulSoup
url = "https://www.python.org/"
response = requests.get(url)

# Extracting the title of the webpage
soup = BeautifulSoup(response.content, 'html.parser')
title = soup.title.string
print(f"Title of the webpage: {title}")
```

```python
filename = "webpage_body.txt"
with open(filename, "w", encoding="utf-8") as f:
    f.write(str(soup.get_text))
```

```python
urls = []
for link in soup.find_all('a'):
    urls.append(link.get('href'))
print(f"All URLs referred to in the webpage: {urls}")
```
