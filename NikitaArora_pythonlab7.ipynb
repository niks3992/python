1. Write a python program to read a dataset from company sales data.

a) Read Total profit of all months and show it using a line plot.

b) Read all product sales data and show it using a multiline plot.

c) Read toothpaste sales data of each month and show it using a
scatter plot.

d) Read sales data of bathing soap of all months and show it using a
bar chart.

e) Calculate total sale data for last year for each product and show it
using a Pie chart.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
ds= pd.read_csv("/content/company-sales.csv")
ds
```

```python
plt.plot(ds['month_number'], ds['total_profit'], color='red')
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
```

```python
plt.plot(ds['month_number'], ds['facecream'], label='Face Cream', marker='o')
plt.plot(ds['month_number'], ds['facewash'], label='Face Wash', marker='o')
plt.plot(ds['month_number'], ds['toothpaste'], label='Toothpaste', marker='o')
plt.plot(ds['month_number'], ds['bathingsoap'], label='Bathing Soap', marker='o')
plt.plot(ds['month_number'], ds['shampoo'], label='Shampoo', marker='o')
plt.plot(ds['month_number'], ds['moisturizer'], label='Moisturizer', marker='o')
plt.title('Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')
```

```python
plt.scatter(ds['month_number'], ds['toothpaste'])
plt.xlabel('Month Number')
plt.ylabel('Toothpaste')
plt.title('Toothpaste Sales Data')
```

```python
plt.bar(ds['month_number'], ds['bathingsoap'],color='blue')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap')
plt.title('Bathing Sales Data')
```

```python
header=["facecream",	"facewash",	"toothpaste",	"bathingsoap",	"shampoo",	"moisturizer"]
totalsale=[ds['facecream'].sum(),ds['facewash'].sum(),ds['toothpaste'].sum(),ds['bathingsoap'].sum(),ds['shampoo'].sum(),ds['moisturizer'].sum()]
plt.pie(totalsale,labels=header,shadow=True,autopct='%1.1f%%',explode = (0.0, 0.0, 0.0, 0.2, 0.0, 0.0))
plt.title('Total Sales Data for all Product')
```
